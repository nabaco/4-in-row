# K-Means algorithm visualization

```python
from models.k_means import KMeans
import numpy as np
from random import choices
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X1 = np.random.rand(100, 2)
X2 = np.random.rand(100, 2) + 0.5
X = np.row_stack((X1, X2))
```

```python
plt.scatter(X[:,0], X[:,1], c="black")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
```

```python
k_means = KMeans(2, 2)
k_means.fit(X, 10, 5)
labels = k_means.predict(X)
centers = k_means.cluster_centers
```

```python
for sample, label in zip(X, labels):
    color = "red" if label == 0 else "blue"
    label = "cluster 0" if label == 0 else "cluster 1"
    plt.scatter(sample[0], sample[1], c=color)
plt.scatter(centers[:, 0], centers[:, 1], c="yellow", label="Cluster Centers")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.legend()
```

```python
print(f"loss: {k_means.loss(X, labels)}")
```
